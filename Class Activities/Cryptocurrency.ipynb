```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
join_df = pd.merge(bitcoin_df, dash_df, on='Date')
join_df.head()
```

```python
# Rename columns so that they are differentiated
join_df = join_df.rename(columns={ 'Open_x':'Bitcoin Opn', 'High_x':'Bitcoin High', 'Low_x':'Bitcoin Low',
                                 'Close_x':'Bitcoin Close', 'Volume_x':'Bitcoin Vol', 'Market Cap_x':'Bitcoin Mkt Cap',
                                  'Open_y':'Dash Opn', 'High_y':'Dash High', 'Low_y':'Dash Low',
                                 'Close_y':'Dash Close', 'Volume_y':'Dash Vol', 'Market Cap_y':'Dash Mkt Cap'})
join_df.head()
```

```python
# Collecting best open for Bitcoin and Dash
bitmax = join_df['Bitcoin Opn'].max()
bitmax

# Collecting best close for Bitcoin and Dash
bitclose = join_df['Bitcoin Close'].max()
bitclose
# Collecting the total volume for Bitcoin and Dash
bitvol = join_df['Bitcoin Vol'].sum()/1000000
dashvol= join_df['Dash Vol'].sum()/1000000
```
